<h1 style="text-align: center; font-family: 'Roboto', sans-serif; font-weight:bold; font-size:40px; margin: 40px 0">03 - EDA : Fusion des données</h1>

### Objectif

Rassembler les données statiques et les données temporelles dans un seul et unique fichier, permettant d'être injectées dans notre modèle prédictif.

### Plan du notebook

- Chargement des fichiers statiques et temporelles
- Regroupement des données par employé
- Création du fichier final  

### Prérequis

- Avoir exécuté les notebooks [EDA - Données Statiques](./01_eda_static_data.ipynb) et [EDA - Feature Engineering Temporel](./02_eda_temporal_and_outliers.ipynb).
- Les fichiers `src/data/pre-processed/static_merged.csv` et `src/data/pre-processed/dynamic_merged.csv` doivent exister.

---
## **1. Importation des bibliothèques nécessaires**

```python
import pandas as pd
```

---
## **2. Chargement des fichiers**

```python
static_data_frame = pd.read_csv("../data/pre-processed/static_merged.csv")
dynamic_data_frame = pd.read_csv("../data/pre-processed/dynamic_merged.csv")
```

---
## **3. Regroupement des données**

```python
merged_data_frame = pd.merge(static_data_frame, dynamic_data_frame, on="employee_id", how="inner", validate="one_to_one")
merged_data_frame = merged_data_frame.drop(columns=["employee_id"])
```

---
## **4. Exportation en CSV**

```python
merged_data_frame.to_csv("../data/processed/merged_data.csv", index=False)
```
