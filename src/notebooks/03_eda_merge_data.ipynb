# 03 - EDA : Fonctionnalité de fusion des données

### Objectif

Rassembler les données statiques et les données temporelles dans un seul et unique fichier, permettant d'être injectées dans notre modèle d'IA.  

### Plan du notebook

1. Chargement des fichiers statiques et temporelles
2. Regroupement des données par employé
3. Création du fichier final  

### Prérequis

- Avoir exécuté les notebooks [EDA - Static data](./01_eda_static_data.ipynb) et [EDA - Temporal and outliers](./02_eda_temporal_and_outliers.ipynb).

---

## 0. Import

```python
import pandas as pd
```

## 1. Chargement des fichiers

```python
static_data_frame = pd.read_csv("../data/pre-processed/static_merged.csv")
dynamic_data_frame = pd.read_csv("../data/pre-processed/dynamic_merged.csv")
```

## 2. Regroupement des données

```python
merged_data_frame = pd.merge(static_data_frame, dynamic_data_frame, on="employee_id", how="inner", validate="one_to_one")
```

## 3. Création du fichier final

```python
merged_data_frame.to_csv("../data/processed/merged_data.csv", index=False)
```
